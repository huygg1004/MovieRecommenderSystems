```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import math 
from pandas.core import shared_docs
import scipy.stats

# visualization
import seaborn as sns

# similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
rating = pd.read_csv("/content/drive/MyDrive/DATA.ML.360/ratings.csv")
movies = pd.read_csv("/content/drive/MyDrive/DATA.ML.360/movies.csv")
```

```python
df = pd.merge(rating, movies, on='movieId', how='inner')
df.head()
```

```python
# Create user-item matrix
matrix_item = df.pivot_table(index='title', columns='userId', values='rating')
matrix_item.head()
```

```python
# Normalize user-item matrix
matrix_item_norm = matrix_item.subtract(matrix_item.mean(axis=1), axis = 0)
matrix_item_norm
```

```python
item_similarity_pearson = matrix_item_norm.T.corr()
item_similarity_pearson.head()
```

```python
df_item_similarity_cosine = pd.DataFrame(cosine_similarity(matrix_item_norm.fillna(0.0000001)))
df_item_similarity_cosine
```

```python
df_item_similarity_cosine.columns = item_similarity_pearson.columns
df_item_similarity_cosine.index = item_similarity_pearson.index
```

```python
df_item_similarity_cosine
```

```python
# Item-based recommendation function
def item_based_rec(picked_userid, number_of_similar_items, number_of_recommendations):
  import operator
  # Movies that the target user has not watched
  picked_userid_unwatched = pd.DataFrame(matrix_item_norm[picked_userid].isna()).reset_index()
  picked_userid_unwatched = picked_userid_unwatched[picked_userid_unwatched[picked_userid]==True]['title'].values.tolist()

  # Movies that the target user has watched
  picked_userid_watched = pd.DataFrame(matrix_item_norm[picked_userid].dropna(axis=0, how='all')\
                            .sort_values(ascending=False))\
                            .reset_index()\
                            .rename(columns={picked_userid:'rating'})
  
  # Dictionary to save the unwatched movie and predicted rating pair
  rating_prediction ={}  

  # Loop through unwatched movies          
  for picked_movie in picked_userid_unwatched: 
    # Calculate the similarity score of the picked movie iwth other movies
    # picked_movie_similarity_score = item_similarity_pearson[[picked_movie]].reset_index().rename(columns={picked_movie:'similarity_score'})
    picked_movie_similarity_score = df_item_similarity_cosine[[picked_movie]].reset_index().rename(columns={picked_movie:'similarity_score'})

    # Rank the similarities between the picked user watched movie and the picked unwatched movie.
    picked_userid_watched_similarity = pd.merge(left=picked_userid_watched, 
                                                right=picked_movie_similarity_score, 
                                                on='title', 
                                                how='inner')\
                                        .sort_values('similarity_score', ascending=False)[:number_of_similar_items]

    # Calculate the predicted rating using weighted average of similarity scores and the ratings from user 1
    # predicted_rating = round(np.average(picked_userid_watched_similarity['rating']),6)
    predicted_rating = round(np.average(picked_userid_watched_similarity['rating'], 
                                        weights=picked_userid_watched_similarity['similarity_score']), 6)
    
    # Save the predicted rating in the dictionary
    rating_prediction[picked_movie] = predicted_rating
    # Return the top recommended movies
  return sorted(rating_prediction.items(), key=operator.itemgetter(1), reverse=True)[:number_of_recommendations]

# Get recommendations
recommended_movie = item_based_rec(picked_userid=1, number_of_similar_items=5, number_of_recommendations =20)
recommended_movie
```

```python
output_list_group_3 = []
user_id_lists = [1,6,7]
for user in user_id_lists:
    recommended_movie = item_based_rec(picked_userid=user, number_of_similar_items=5, number_of_recommendations =20)
    output_list_group_3.append(recommended_movie)
output_list_group_3
```

# Granularity Atomic Case in order to answer questions such as Why Not Matrix:
For example: Why Not Matrix:
General Cases:
1. movie name does not exist in database
2. movie name has the same score as another item
3. you asked for few items

Model Based Cases:
1. Too few of your peers have rated this movie (if num of peers <3)
2. Noone has rated this movie
3. All users have given this movie a low score (threshold score = rating of 2 * num of users ratings)

```python
def why_not(movie_name):
  #Why Not ABCXYZ:

  #General:

  tie = False
  if movie_name not in movies['title'].values:
    print("item " + movie_name + " does not exists in the database")
  if movie_name in group_movies_df["movie"].values:
    index_list = group_movies_df.index[group_movies_df['movie'] == movie_name].tolist()
    index_value = index_list[0]
    duplicate_result = group_movies_df["score"].duplicated()
    if duplicate_result[index_value] == True:
      print(movie_name + " has the same score as another item")
      tie = True
    if movie_name not in df_top_20_movies_for_group_of_3.index.values and tie == False:
      print("you asked for few items")
  #Model Based:

  #case: not many people or no one rated the movie

  movie_row = matrix_item.loc[[movie_name]]
  movie_row =  movie_row.dropna(axis=1, how='all')
  count_of_peers = movie_row.shape[1]
  if count_of_peers < 3:
    print("Too few of your peers have rated this movie")
  if count_of_peers == 0:
    print("Noone has rated this movie")
  
  #the number of ratings from users too low
  total = movie_row.sum(axis=1)
  threshold = 2 * movie_row.shape[1]
  if total < threshold:
    print("All users have given this movie a low score")
```

# Position Absenteeism Case for question such as Why not rank Matrix first
1. Your peers have given a low score to this item (if the total score less than threshold of rating of 2 times number of peers in group)
2. Peer X Y Z do not like that movie (if the rating for one peer is less than 2)

```python
user_id_lists = [1,6,7]

def why_not_higher(movie_name, peer1, peer2, peer3):
  movie_row = matrix_item.loc[[movie_name]]
  movie_row =  movie_row.dropna(axis=1, how='all')
  rating1 = movie_row.loc[movie_name][peer1]
  rating2 = movie_row.loc[movie_name][peer2]
  rating3 = movie_row.loc[movie_name][peer3]
  total_rating = rating1 + rating2 + rating3
  threshold = 2 * 3
  if total_rating < threshold:
    print("your peers have given a low score to this item")
  else:
    print("Your peers love this movie")

  output_peer_string = ""
  if(rating1 < 2 or rating2 < 2 or rating3 < 2):
    if rating1 < 2:
      output_peer_string += "Peer " + peer1 + " "
    if rating2 < 2:
      output_peer_string += "Peer " + peer2 + " "
    if rating3 < 2:
      output_peer_string += "Peer " + peer3 + " "
    print(output_peer_string + " don't like that movie")


```

# Granularity Group case to answer questions such as Why No Comedies
2 cases:
1. Your peers dislike that genre (if none of the movies in recommended movies list for group has that genre)
2. Only x of your peers like that genre (if less than 5 peers like that movie of that genre)


```python
def why_no_genre(genre):
  genre_list = rec_with_genres['genres'].tolist()
  genre_substring = genre
  found = 0
  for genres in genre_list:
    if genre_substring in genres:
      found = found + 1
  if found == 0:
    print("Your peers dislike " + genre)
  elif found < 5:
    print("Only " + str(found) + " of your peers like " + genre)
```
